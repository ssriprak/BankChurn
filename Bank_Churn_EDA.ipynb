```python
# Bank Customer Churn - EDA Notebook

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Churn Modeling.csv')
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)

# Set style for plots
sns.set(style="whitegrid")
```

-------------------------
## Univariate Analysis
-------------------------


```python
# Churn distribution
sns.countplot(data=df, x='Exited')
plt.title('Churn Distribution')
plt.xlabel('Exited (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python

# Gender distribution
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')
plt.show()
```

```python
# Geography distribution
sns.countplot(data=df, x='Geography')
plt.title('Geography Distribution')
plt.show()
```

```python
# Age distribution
sns.histplot(data=df, x='Age', kde=True, bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
# Balance distribution
sns.histplot(data=df, x='Balance', kde=True, bins=30)
plt.title('Balance Distribution')
plt.show()
```

```python
# Estimated Salary distribution
sns.histplot(data=df, x='EstimatedSalary', kde=True, bins=30)
plt.title('Estimated Salary Distribution')
plt.show()
```

-------------------------
## Bivariate Analysis
-------------------------


```python
# Churn by Gender
sns.countplot(data=df, x='Gender', hue='Exited')
plt.title('Churn by Gender')
plt.show()
```

```python
# Churn by Geography
sns.countplot(data=df, x='Geography', hue='Exited')
plt.title('Churn by Geography')
plt.show()
```

```python
# Churn by Tenure
sns.countplot(data=df, x='Tenure', hue='Exited')
plt.title('Churn by Tenure')
plt.show()
```

```python
# Churn by NumOfProducts
sns.countplot(data=df, x='NumOfProducts', hue='Exited')
plt.title('Churn by Number of Products')
plt.show()
```

```python
# Churn by HasCrCard
sns.countplot(data=df, x='HasCrCard', hue='Exited')
plt.title('Churn by Credit Card Ownership')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Churn by IsActiveMember
sns.countplot(data=df, x='IsActiveMember', hue='Exited')
plt.title('Churn by Active Membership')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

-------------------------
## Boxplots
-------------------------

```python

# Balance vs Churn
sns.boxplot(data=df, x='Exited', y='Balance')
plt.title('Balance vs Churn')
plt.show()

# Age vs Churn
sns.boxplot(data=df, x='Exited', y='Age')
plt.title('Age vs Churn')
plt.show()

# Estimated Salary vs Churn
sns.boxplot(data=df, x='Exited', y='EstimatedSalary')
plt.title('Estimated Salary vs Churn')
plt.show()
```

-------------------------
## KDE Plots
-------------------------


```python

# KDE of Age by Churn
sns.kdeplot(data=df, x='Age', hue='Exited', fill=True)
plt.title('Age Distribution by Churn')
plt.show()

# KDE of Balance by Churn
sns.kdeplot(data=df, x='Balance', hue='Exited', fill=True)
plt.title('Balance Distribution by Churn')
plt.show()

# KDE of EstimatedSalary by Churn
sns.kdeplot(data=df, x='EstimatedSalary', hue='Exited', fill=True)
plt.title('Estimated Salary Distribution by Churn')
plt.show()
```

```python
# -------------------------
# Correlation Heatmap
# -------------------------
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
